```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('dog.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img_circle=img_rgb.copy()
img_line=img_rgb.copy()
img_rectangle=img_rgb.copy()
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
cv2.circle(img_circle, (140, 120), 40, (0, 0, 255), -1)
plt.imshow(img_circle)
plt.show()
```

```python
cv2.line(img_line, (70, 50), (100, 50), (255, 0, 0), 3)
plt.imshow(img_line)
plt.show()
```

```python
cv2.rectangle(img_rectangle, (60, 60), (160, 160), (0, 255, 0), 3)
plt.imshow(img_rectangle)
plt.show()
```

```python
cv2.line(img_rgb, (70, 50), (100, 50), (255, 0, 0), 3)
cv2.rectangle(img_rgb, (60, 60), (160, 160), (0, 255, 0), 3)
cv2.circle(img_rgb, (140, 120), 40, (0, 0, 255), -1)
plt.imshow(img_rgb)
plt.show()
```
